```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_category(csv_file, title, out_file):
    data = pd.read_csv(csv_file)

    plt.figure(figsize=(10, 6))
    plt.plot(data['n'], data['quick_us'], label='Quick Sort (Random Pivot)', linewidth=1.5)
    plt.plot(data['n'], data['hybrid_us'], label='Hybrid Quick+Heap+Insertion (Introsort)', linewidth=1.5)

    plt.xlabel("  n")
    plt.ylabel(" ()")
    plt.title(title)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig(out_file, dpi=300)
    plt.show()
    plt.close()


plot_category(
    "quick_random.csv",
    "     ",
    "plot_quick_random.png"
)

plot_category(
    "quick_reversed.csv",
    "    ,   ",
    "plot_quick_reversed.png"
)

plot_category(
    "quick_almost.csv",
    "      ",
    "plot_quick_almost.png"
)

```
