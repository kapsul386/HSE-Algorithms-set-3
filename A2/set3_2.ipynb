```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_category(csv_file, title, out_file):
    data = pd.read_csv(csv_file)

    plt.figure(figsize=(10, 6))
    plt.plot(data['n'], data['merge_us'], label='Merge Sort', linewidth=1.5)
    plt.plot(data['n'], data['hybrid_us'], label='Hybrid Merge+Insertion', linewidth=1.5)

    plt.xlabel("Размер массива n")
    plt.ylabel("Время (микросекунды)")
    plt.title(title)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig(out_file, dpi=300)
    plt.show()
    plt.close()


plot_category(
    "random_arrays.csv",
    "Время работы сортировок на случайных массивах",
    "plot_random.png"
)

plot_category(
    "reversed_arrays.csv",
    "Время работы сортировок на массивах, отсортированных по убыванию",
    "plot_reversed.png"
)

plot_category(
    "almost_sorted_arrays.csv",
    "Время работы сортировок на почти отсортированных массивах",
    "plot_almost.png"
)

print("Графики сохранены: plot_random.png, plot_reversed.png, plot_almost.png")
```

На всех трёх группах тестовых данных - случайных, почти отсортированных и отсортированных в обратном порядке - гибридная реализация MERGE+INSERTION SORT уверенно превосходит классический MERGE SORT по времени работы. Для каждого размера массива кривая гибридного алгоритма располагается заметно ниже, чем у стандартного MERGE SORT, а рост времени происходит более плавно.

Особенно выраженный выигрыш наблюдается на почти отсортированных массивах: здесь вставочная сортировка внутри гибридного алгоритма работает наиболее эффективно, что приводит к наибольшему ускорению. На случайных и обратно отсортированных данных преимущества также сохраняются для всех диапазонов n, причём разница становится особенно заметной на средних и крупных размерах массива.

В целом эксперимент подтверждает, что использование вставочной сортировки на малых подмассивах снижает накладные расходы рекурсии и позволяет гибридному алгоритму работать быстрее стандартного MERGE SORT практически на всех рассматриваемых входах.
