Нужно запустить код снизу

```python
import math
import pandas as pd
import matplotlib.pyplot as plt

S_exact = 0.25 * math.pi + 1.25 * math.asin(0.8) - 1.0

areas = pd.read_csv("areas.csv")
errors = pd.read_csv("relative_errors.csv")


plt.figure()
plt.plot(areas["N"], areas["area_wide"], label="Широкий прямоугольник")
plt.plot(areas["N"], areas["area_narrow"], label="Узкий прямоугольник")
plt.axhline(S_exact, linestyle="--", label="Точная площадь")

plt.xlabel("Число точек N")
plt.ylabel("Оценка площади")
plt.title("Оценка площади пересечения трёх кругов методом Монте–Карло")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("areas_plot.png", dpi=300)


plt.figure()
plt.plot(errors["N"], errors["rel_error_wide"], label="Широкий прямоугольник")
plt.plot(errors["N"], errors["rel_error_narrow"], label="Узкий прямоугольник")
plt.yscale("log")

plt.xlabel("Число точек N")
plt.ylabel("Относительная ошибка")
plt.title("Относительная ошибка оценки площади")
plt.grid(True, which="both")
plt.legend()
plt.tight_layout()
plt.savefig("relative_errors_plot.png", dpi=300)

plt.show()
```

По графикам видно, что при одинаковом числе точек узкая прямоугольная область даёт существенно более стабильные и точные оценки площади. Относительная ошибка при использовании узкой области заметно ниже и быстрее убывает с ростом числа точек.
